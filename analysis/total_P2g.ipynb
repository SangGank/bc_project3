```python
from g2pk import G2p
import pandas as pd

g2p = G2p()

train = pd.read_csv('../argument_data/train_notNoise.csv')
train['type']=0

train_per = train.copy()
train_des = train.copy()
train_per.type=1
train_des.type=2
train.to_csv('../argument_data/train_notNoise_type.csv', index =False)

train_per.text= train_per.text.apply(lambda x: g2p(x))
train_per.to_csv('../argument_data/train_notNoise_per.csv', index =False)

train_des.text= train_des.text.apply(lambda x: g2p(x,descriptive=True))
train_des.to_csv('../argument_data/train_notNoise_des.csv',index =False)
```
