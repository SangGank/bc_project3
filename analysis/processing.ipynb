```python
import pandas as pd
import re
```

```python
train_data = pd.read_csv('../data/train2.csv')
train_data.head()
```

```python
def processing(text):
    text = re.sub(r'<[^>]+\s+(?=<)|<[^>]+>',"",text)
    text= re.sub(r'(amp;|lt;|gt;)','',text)
    text= re.sub(r'[,&\[\]\'\"\-\(\)\~\+\?#:\/!]','',text).strip()
    return text
```

```python
train_data2=train_data.copy()
train_data2.real = train_data2.real.apply(lambda x: processing(x))
```

```python
train_data2[train_data2.real != train_data2.text][['text','real']]
```
